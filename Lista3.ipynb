Aprendizado por reforço é uma técnica de aprendizado de máquina que se baseia em recompensas e punições para guiar o modelo em direção ao comportamento desejado. Em jogos, isso pode ser usado para treinar um agente para jogar o jogo de maneira eficiente. O agente é treinado para tomar ações que maximizam a recompensa (como ganhar pontos no jogo) e evitar ações que resultam em punição (como perder vidas no jogo). Ao continuar a experimentar o jogo e receber feedback sobre suas ações, o agente aprende a tomar decisões mais eficientes e a jogar o jogo com habilidade cada vez maior.


image.png

. Em jogos RTS, o jogador controla uma equipe de unidades e deve tomar decisões estratégicas em tempo real para vencer o oponente. Treinar um agente para jogar um jogo RTS usando aprendizado por reforço envolve definir recompensas e punições apropriadas para diferentes ações no jogo, como construir unidades ou atacar o oponente. O agente é então treinado a tomar decisões baseadas nessas recompensas e punições, a fim de maximizar sua pontuação e vencer o jogo. Como no caso de outros jogos, o agente continuará a experimentar o jogo e receber feedback sobre suas ações, o que lhe permitirá aprender a jogar cada vez melhor.

# Funcionamento desse jogo 
O jogo consiste de dois jogadores disputando por recursos, construindo unidades e as controlando
com o objetivo final de destruir a base do adversário. Mais precisamente, cada jogador começa com
sua base em cantos opostos de um mapa quadricular, que contém fontes de recursos. É importante
ressaltar que os jogadores só tem visão do que está suficiente próximo de suas unidades (e conhecimento da localização da base inimiga), isso significa que se trata de um problema de informação
parcial, ilustrado na figura. Assim, a formulação do problema como um MDP é uma aproximação.

Ele funciona treinando o modelo em um ambiente em que ele pode tomar ações e receber feedback sobre suas decisões. O modelo é inicialmente inicializado com uma política (ou seja, um conjunto de regras para tomar decisões) e então é exposto ao ambiente. Quando o modelo toma uma ação, ele recebe uma recompensa ou uma punição de acordo com o resultado dessa ação. O modelo usa essas recompensas e punições para atualizar sua política de maneira a maximizar a recompensa total ao longo do tempo. Como resultado, o modelo aprende a tomar decisões mais eficientes e a produzir comportamentos mais desejáveis.

##Algoritmo

Atacaremos o problema definindo uma parametrização da política, ou seja, uma família de funções
da forma πθ
(a|s) = P(At = a|St = s, θt = θ) e uma função objetivo J (θ) de tal forma que nosso
problema se reduz a encontrar um vetor θ ∈ R
d que maximiza a função J . Nossa função objetivo
será escolhida como J (θ)
.
= vπθ
(S0
), ou seja, o valor do estado inicial na política definida por θ.
Essa função é diferenciável e além disso temos uma forma conveniente de expressar seu gradiente
através do teorema do gradiente de política [2]:

image.png

Esse algoritmo é próximo da teoria porém é relativamente lento na prática, principalmente pela
variância no processo de treinamento, dado que as atualizações dependem totalmente dos episódios
observados. Daí surge a classe de algoritmos Actor-Critic [2], que utilizam uma estimativa do valor
do estado ao invés do retorno, diminuindo a variância no treinamento. O algoritmo A3C, de Asynchronous Advantage Actor-Critic [1], é uma modificação do Actor-Critic que usa fortemente paralelismo
para diversificar a experiência do agente a cada atualização dos parâmetros (o que deixa o treinamento
mais estável) e usufruir totalmente de processadores multi-core, por esses motivos, o treinamento com
o algoritmo A3C tende a ser significativamente mais rápido do que com Actor-Critic puro.

```python
import numpy

# Inicialização dos parâmetros
GENES = 2
# Definimos a taxa de crossover (reprodução)
CROSSOVER_RATE = 15
mattingPoolSize = 6
offspringSize = CROSSOVER_RATE - mattingPoolSize
# Definimos o limite superior e inferior para os valores de x
LOWER_LIMIT = -1
UPPER_LIMIT = 2
# Definimos o tamanho da população
POPULATION_SIZE = (CROSSOVER_RATE, GENES)
# Definimos o número de gerações
NUM_GENERATIONS = 3

# Definimos a taxa de mutação
MUTATION_RATE = 3


population = numpy.random.uniform(LOWER_LIMIT, UPPER_LIMIT, POPULATION_SIZE)

for generation in range(NUM_GENERATIONS):
    print(("Generation:", generation+1))
    fitness = population*numpy.sin(10*np.pi*population) + 1
    parents = numpy.empty((mattingPoolSize, population.shape[1]))
    for p in range(mattingPoolSize):
        fittestIndex = numpy.where(fitness == numpy.max(fitness))
        fittestIndex = fittestIndex[0][0]
        parents[p, :] = population[fittestIndex, :]
        fitness[fittestIndex] = -1
    offspring = numpy.empty((offspringSize, population.shape[1]))
    for k in range(offspringSize):
         crossoverPoint = numpy.random.randint(0,GENES) 
         parent1Index = k%parents.shape[0]
         parent2Index = (k+1)%parents.shape[0]
         offspring[k, 0: crossoverPoint] = parents[parent1Index, 0: crossoverPoint]
         offspring[k, crossoverPoint:] = parents[parent2Index, crossoverPoint:]
    for index in range(offspring.shape[0]):
        randomIndex = numpy.random.randint(1,GENES)
        randomValue = numpy.random.uniform(LOWER_LIMIT, UPPER_LIMIT, 1)
        offspring [index, randomIndex] = offspring [index, randomIndex] + randomValue

    population[0:parents.shape[0], :] = parents
    population[parents.shape[0]:, :] = offspring
    print("\n Próxima população")
    print(population)

fitness = population*numpy.sin(10*np.pi*population) + 1
fittestIndex = numpy.where(fitness == numpy.max(fitness))
fittestIndex = fittestIndex[0][0]
fittestInd = population[fittestIndex, :]
bestFitness = fitness[fittestIndex]

print("\nMelhor Indivíduo:")
print(fittestInd)
print("\nMelhor Aptidão:")
print(bestFitness)
```
