# Funcionamento desse jogo 
O jogo consiste de dois jogadores disputando por recursos, construindo unidades e as controlando
com o objetivo final de destruir a base do adversrio. Mais precisamente, cada jogador comea com
sua base em cantos opostos de um mapa quadricular, que contm fontes de recursos.  importante
ressaltar que os jogadores s tem viso do que est suficiente prximo de suas unidades (e conhecimento da localizao da base inimiga), isso significa que se trata de um problema de informao
parcial, ilustrado na figura. Assim, a formulao do problema como um MDP  uma aproximao.

Ele funciona treinando o modelo em um ambiente em que ele pode tomar aes e receber feedback sobre suas decises. O modelo  inicialmente inicializado com uma poltica (ou seja, um conjunto de regras para tomar decises) e ento  exposto ao ambiente. Quando o modelo toma uma ao, ele recebe uma recompensa ou uma punio de acordo com o resultado dessa ao. O modelo usa essas recompensas e punies para atualizar sua poltica de maneira a maximizar a recompensa total ao longo do tempo. Como resultado, o modelo aprende a tomar decises mais eficientes e a produzir comportamentos mais desejveis.

```python
import numpy

# Inicializao dos parmetros
GENES = 2
# Definimos a taxa de crossover (reproduo)
CROSSOVER_RATE = 15
mattingPoolSize = 6
offspringSize = CROSSOVER_RATE - mattingPoolSize
# Definimos o limite superior e inferior para os valores de x
LOWER_LIMIT = -1
UPPER_LIMIT = 2
# Definimos o tamanho da populao
POPULATION_SIZE = (CROSSOVER_RATE, GENES)
# Definimos o nmero de geraes
NUM_GENERATIONS = 3

# Definimos a taxa de mutao
MUTATION_RATE = 3


population = numpy.random.uniform(LOWER_LIMIT, UPPER_LIMIT, POPULATION_SIZE)

for generation in range(NUM_GENERATIONS):
    print(("Generation:", generation+1))
    fitness = population*numpy.sin(10*np.pi*population) + 1
    parents = numpy.empty((mattingPoolSize, population.shape[1]))
    for p in range(mattingPoolSize):
        fittestIndex = numpy.where(fitness == numpy.max(fitness))
        fittestIndex = fittestIndex[0][0]
        parents[p, :] = population[fittestIndex, :]
        fitness[fittestIndex] = -1
    offspring = numpy.empty((offspringSize, population.shape[1]))
    for k in range(offspringSize):
         crossoverPoint = numpy.random.randint(0,GENES) 
         parent1Index = k%parents.shape[0]
         parent2Index = (k+1)%parents.shape[0]
         offspring[k, 0: crossoverPoint] = parents[parent1Index, 0: crossoverPoint]
         offspring[k, crossoverPoint:] = parents[parent2Index, crossoverPoint:]
    for index in range(offspring.shape[0]):
        randomIndex = numpy.random.randint(1,GENES)
        randomValue = numpy.random.uniform(LOWER_LIMIT, UPPER_LIMIT, 1)
        offspring [index, randomIndex] = offspring [index, randomIndex] + randomValue

    population[0:parents.shape[0], :] = parents
    population[parents.shape[0]:, :] = offspring
    print("\n Prxima populao")
    print(population)

fitness = population*numpy.sin(10*np.pi*population) + 1
fittestIndex = numpy.where(fitness == numpy.max(fitness))
fittestIndex = fittestIndex[0][0]
fittestInd = population[fittestIndex, :]
bestFitness = fitness[fittestIndex]

print("\nMelhor Indivduo:")
print(fittestInd)
print("\nMelhor Aptido:")
print(bestFitness)
```
